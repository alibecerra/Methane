```python
import pandas as pd

df3 = pd.read_csv("/home/aj/cojects/Methane/data/plumes_TX.csv", on_bad_lines='skip')
#print(df.head())
#print(df.describe())
print(df3.columns.to_list())
```

```python
import folium

# Create a map centered around major emission sites
m = folium.Map(location=[df3["plume_latitude"].mean(), df3["plume_longitude"].mean()], zoom_start=6)

# Add emission sites as markers
for _, row in df3.iterrows():
    folium.CircleMarker(
        location=[row["plume_latitude"], row["plume_longitude"]],
        radius=5,
        color="red",
        fill=True,
        fill_color="blue",
    ).add_to(m)

# Show the map
m
```
